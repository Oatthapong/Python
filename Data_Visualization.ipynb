# นำเข้า Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# สร้างข้อมูลตัวอย่าง
data = {
    "customer_id": [101, 102, 103, 104, 105, 106],
    "age": [25, 30, 35, 40, 45, 50],
    "purchase_amount": [10000, 15000, 20000, 25000, 30000, 35000],
    "gender": ["F", "M", "M", "M", "F", "M"]
}

df = pd.DataFrame(data)
print(df)
```

# แผนภูมิแท่ง (Bar Chart) - เปรียบเทียบยอดซื้อของแต่ละเพศ

```python
plt.figure(figsize=(6,4))

# คำนวณยอดซื้อเฉลี่ยตามเพศ
df_grouped = df.groupby("gender")["purchase_amount"].mean()

# สร้าง Bar Chart
df_grouped.plot(kind="bar", color=["pink", "blue"])
plt.xlabel("Gender")
plt.ylabel("Average Purchase Amount")
plt.title("Average Purchase Amount by Gender")
plt.xticks(rotation=0)
plt.show()
```

# ฮิสโตแกรม (Histogram) - กระจายตัวของอายุ

```python
plt.figure(figsize=(6,4))
plt.hist(df["age"], bins=5, color="skyblue", edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Customer Ages")
plt.show()
```

# กราฟเส้น (Line Chart) - แนวโน้มยอดซื้อ

```python
plt.figure(figsize=(6,4))
plt.plot(df["age"], df["purchase_amount"], marker="o", linestyle="-", color="red")
plt.xlabel("Age")
plt.ylabel("Purchase Amount")
plt.title("Purchase Trend by Age")
plt.grid()
plt.show()
```

# กราฟกระจาย (Scatter Plot) - ความสัมพันธ์ระหว่างอายุกับยอดซื้อ

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["age"], y=df["purchase_amount"], hue=df["gender"], palette={"F": "pink", "M": "blue"})
plt.xlabel("Age")
plt.ylabel("Purchase Amount")
plt.title("Age vs Purchase Amount")
plt.show()
```
